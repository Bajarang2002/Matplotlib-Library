```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("Top100moststreamed.csv")
df.head(10)
```

```python
q1=df['year'].value_counts()
q1
```

```python
q1 =df['popularity'].value_counts()
q1 =q1[q1>1]
q1
```

```python
print(q1.index)
print(q1.values)
```

```python
plt.plot(q1.index,q1.values,color ='red',label='count',marker ='*')
plt.xlabel('year')
plt.ylabel('popularity')
plt.title('Yearwise popularity',fontsize = 15,color='blue')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,4))
plt.plot(q1.index,q1.values,color ='green',label='count',marker ='o',markersize =15)
plt.xlabel('year')
plt.ylabel('popularity')
plt.title('Yearwise popularity',fontsize = 15,color='blue')
plt.grid()
plt.savefig('Lineplot.png')
plt.legend()
plt.show()
```

```python
plt.plot(q1.index,q1.values,color ='maroon',label='count',lw =3,linestyle=':',marker ='*')
plt.xlabel('year')
plt.ylabel('popularity')
plt.title('Yearwise popularity',fontsize = 15,color='blue')
plt.grid()
plt.legend()
plt.show()
```

```python
# Scatter

q2 = df.groupby('year')['acousticness'].mean().reset_index()
q3=q2.sort_values(by ='acousticness',ascending True)
q3.head()
```

```python
plt.scatter(q1.index,q1.values,color ='red',label='count',marker ='*',s =15)
plt.xlabel('year')
plt.ylabel('popularity')
plt.title('year vs popularity',fontsize =15,color ='blue')
plt.grid()
plt.legend()
plt.show()
```
