```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("Top100moststreamed.csv")
df.head(10)
```

```python
df.columns
```

```python
df.iloc[:11,:16]
```

```python
q1 = df.groupby("year")["energy"].mean().reset_index()
q1 =q1.sort_values(by ='energy',ascending=True)
q1.head()
```

```python
plt.bar(q1['year'],q1['energy'],color='blue',label ='red')
plt.title('Year wise average energy',fontsize=15,color ='orange')
plt.xlabel('Title year')
plt.ylabel('energy')
plt.legend()
plt.show()
```

```python
plt.bar(q1['year'],q1['energy'],color=['Red','orange','blue','purple','pink','yellow','green'],edgecolor ='black')
plt.title('Year wise average energy',fontsize=15,color ='orange')
plt.xlabel('Title year',color ='violet')
plt.ylabel('energy')
plt.legend()
plt.show()
```

```python
q1 = df.groupby('top genre')[['year','valance']].mean()
q1.index.names = ['genre']
q1.head()
```

```python
plt.bar(q1.index,q1['year'],color ='red',label = 'year',edgecolor ='black')   
plt.bar(q1.index,q1['valance'],color ='maroon',label = 'valance',edgecolor ='black') 
plt.xlabel('year')
plt.xticks(rotation =90)
plt.title('Yearwise valance',fontsize =15,color ='blue')
plt.ylabel('valance')
plt.legend(loc =1)
plt.grid()
plt.show()
```

```python
q2 = df.groupby('artist')['popularity'].mean().reset_index()
q2 = q2.sort_values(by='popularity',ascending = False)
q2.head()
```

```python
plt.bar(q2['artist'],q2['popularity'],color ='yellow',label = 'popularity',edgecolor ='black') 
plt.xlabel('artist')
plt.xticks(rotation =45)
plt.title('Artist and Popularity',fontsize =15,color ='blue')
plt.ylabel('popularity')
plt.legend(loc =1)
plt.grid()
```

```python
plt.figure(figsize=(8,5))
plt.bar(q2['artist'],q2['popularity'],color ='yellow',label = 'popularity',edgecolor ='black') 
plt.xlabel('artist')
plt.xticks(rotation =90)
plt.title('Artist and Popularity',fontsize =15,color ='blue')
plt.ylabel('popularity')
plt.ylim(0,500)
plt.legend(loc =1)
plt.savefig('artist and popuarity.png')
plt.grid()
```
